# image to pencil sketch with python

```python

from google.colab.patches import cv2_imshow
```

```python

import cv2
```

```python
pic = cv2.imread("sayginyalcin.jpg")
```

```python
cv2_imshow(pic)
```

One of the best Dubai-based German entrepreneur** Mr.Saygin Yalcin**

```python
pic_gray = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(pic_gray)
```

```python
pic_invert = cv2.bitwise_not(pic_gray)
```

```python
cv2_imshow(pic_invert)

```

```python
pic_smoothing = cv2.GaussianBlur(pic_invert, (21, 21),0)
```

```python
cv2_imshow(pic_smoothing)
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
    
```

```python
finishing_pic = dodgeV2(pic_gray, pic_smoothing)
```

```python
cv2_imshow(finishing_pic)
```
